# Splitting data intrain and test 

```python
new_df.columns
```

```python
X= new_df[['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']]
X
```

```python
y= new_df['Class']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=15)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Model Training 

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
x_train_pred =  model.predict(X_train)
x_train_pred
```

```python
train_model_accuracy = accuracy_score(x_train_pred, y_train)*100
print('Train_model accuraxcy is -', train_model_accuracy)
```

```python
X_test_pred = model.predict(X_test)
```

```python
X_test_pred
```

```python
model_accuracy = accuracy_score(X_test_pred, y_test)*100
print('Test_model accuraxcy is -', model_accuracy)
```
